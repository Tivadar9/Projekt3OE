```python
import csv
import pandas as pd
import pyspark;
from pyspark.sql.functions import year, current_date
from pyspark.sql import SparkSession
from pyspark.sql.functions import sum, avg

base_path = "/Workspace/Users/viktorkapitany@stud.uni-obuda.hu/Projekt3OE"

df = []
with open(base_path + "/cars.txt", "r") as f:
    reader = csv.reader(f, delimiter=';')
    header = next(reader)
    for row in reader:
        df.append(row)
df = pd.DataFrame(df, columns=header)
#display(df)

#Transzformációk
df.drop(df.columns[5], axis=1, inplace=True) #Üres oszlop törlése
df["Year"] = df["Year"].astype(int)
df["Registered"] = df["Registered"].astype(bool)

df["Age"] = pd.Timestamp.today().year - df["Year"]

#Aggregált adatok
df_agg = df.groupby(["Manufacturer", "Type", "FuelType"]).agg({"Age": "mean"}).reset_index()
df_agg = df.sort_values(by=["Manufacturer", "Type", "FuelType"])
#display(df_agg)

dfsum = df.groupby("Manufacturer").agg({"Year": ["sum", "mean"]}).reset_index()
dfsum.columns = ["Manufacturer", "sum_year", "avg_year"]
dfsum["avg_year"] = dfsum["avg_year"].round(1)

display(dfsum)
```
