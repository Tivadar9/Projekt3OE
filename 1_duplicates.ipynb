```python
# JSON fájlok betöltése egy DataFrame-be
import pandas as pd
import json

base_path = "/Workspace/Users/szab.krizosztom@stud.uni-obuda.hu/Projekt3OE/"
files = [
    base_path + "airports_with_duplicates.json",
    base_path + "car_sales_with_duplicates.json",
    base_path + "customer_with_duplicates.json",
    base_path + "electricity_with_duplicates.json"
]

# Load JSON files using pandas instead of Spark
dfs = []
for file in files:
    with open(file, 'r') as f:
        data = json.load(f)
    dfs.append(pd.DataFrame(data))

df = pd.concat(dfs, ignore_index=True)

print(f" records: {len(df)}")
```

```python
# Duplikált rekordok elkülönítése
dedup_df = df.drop_duplicates()
dup_df = df[df.duplicated(keep=False)]
```

```python
# Adatok visszaírása JSON fájlokba
base_path = "/Workspace/Users/szab.krizosztom@stud.uni-obuda.hu/Projekt3OE/"

dedup_df.to_json(base_path + "deduplicate.json", orient="records", indent=2)
dup_df.to_json(base_path + "duplicate.json", orient="records", indent=2)

print(f"Deduplicated records: {len(dedup_df)}")
print(f"Duplicate records: {len(dup_df)}")
```
